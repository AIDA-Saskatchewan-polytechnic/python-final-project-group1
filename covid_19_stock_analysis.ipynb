#### Prepare Americal Stock List From AlphaVantage API

```python
df_usa_stocks = pd.read_csv('stock_data/american_stocks.csv',header=0)
df_usa_stocks

base_url = 'https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol='
api_key = '&apikey=' + 'P5R3380DQ77VJ1VM'

usa_stock_list=[]

usa_stock_json = {
    "Name":[],
    "Symbol":[],
    "Details":[]
}


date_column_list = df_conf_and_death.columns.values.tolist()
date_columns = date_column_list[4:]
date_columns.insert(0,"Name")
date_columns.insert(1,"Symbol")


usa_stock_df_json = {}

# AlphaVantage API to get stock details
for stock_name, stock in zip(df_usa_stocks['Name'],df_usa_stocks['Symbol']):    
    url = base_url + stock + api_key
    resp = requests.get(url)
    data = resp.json()
    usa_stock_df_json[]
    usa_stock_df_json['Name'].append(stock_name)
    usa_stock_df_json['Symbol'].append(stock)
    stock_details = data['Time Series (Daily)']

    for stock_detail in stock_details:
        usa_stock_df_json['Date'].append(stock_detail)
        usa_stock_df_json['Hight'].append(stock_details[stock_detail]['1. open'])
        usa_stock_df_json['Low'].append(stock_details[stock_detail]['2. high'])
        usa_stock_df_json['Open'].append(stock_details[stock_detail]['3. low'])
        usa_stock_df_json['Close'].append(stock_details[stock_detail]['4. close'])
        usa_stock_df_json['Volume'].append(stock_details[stock_detail]['5. volume'])



print(usa_stock_df_json)
    

```
